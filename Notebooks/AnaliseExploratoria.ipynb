# *DNA*

Provavelmente você já viu objetos similares aos da imagem acima em algum momento da sua vida. Se na sua cabeça veio a intuição de que ela representa o DNA, você está certo!

## Mas o que é DNA?

Podemos pensar o DNA como uma bolsa que carrega todas as ferramentas necessárias para que o conjunto de funções de um ser vivo, isto é, seu organismo, nasça e se mantenha. E é justamente por isso que o DNA muitas vezes é caacterizado como a chave da vida, já que além de sua importância pelo que foi explicado acima, ele está presente em todo ser vivo: ele é o elemento comum em toda vida na Terra.



# *Genes*

Pedaços do DNA são conhecidos como genes. No ser humano, cada gene tem a função de ajudar na produção de proteínas de forma equilibrada para manter o bom funcionamento do organismo. As expressões gênicas são medidas que nos informam sobre a quantidade dessa produção: ela pode aumentar, diminuir ou se manter estável. E é justamente pra fazer essa manipulação que a maioria dos remédios são desenvolvidos.

Nesse sentido, quanto maior o valor de expressão genética de um gene, mais determinada proteína ele está ajudando a produzir e vice versa.

# Células

maquinas.jpg

Fonte: [Wikipédia](https://pt.wikipedia.org/wiki/Ind%C3%BAstria_t%C3%AAxtil)

Na imagem acima temos várias máquinas industriais de fabricação de tecidos. Na foto, enxergamos somente aquilo que é maior, os elementos externos das máquinas. Mas é claro que cada uma delas possui diversas peças, muitas bem pequenas, que são responsáveis pelo seu funcionamento. As células, fazendo uma analogia com o exemplo dado, seriam os parafusos (peças menores) dos seres humanos.

Menores elementos presentes em nosso organismo, elas são responsáveis por diversas funções e conseguem se multiplicar a partir da quantidade de expressão gênica explicada mais acima. As células possuem, quando agindo de forma esperada: 

a) uma função; 

b) um tempo para nascer; 

c) um tempo para realizar sua função e 

d) o momento de morrer.

Porém, podem acontecer desordens no organismo causado por alguma anormalidade em algum desses aspectos das células. Células cancerígenas, por exemplo, ocasionam a multiplicação desordenada das células espalhando-as por todo o corpo. Aproveitando nossa analogia com as máquinas de tecidos, imagine vários parafusos se multiplicando dentro de uma das máquinas e, dentre outros problemas, emperrando a engrenagem, por exemplo.

Um composto químico, que pode vir a se tornar um remédio, afeta tanto as expressões gênicas quanto, por causa disto, a multiplicação ou morte dos vários tipos de células presente em nosso corpo.

 Para ficar mais claro, podemos exemplificar com pacientes diabéticos e a proteína insulina.

A insulina é responsável por diminuir a quantidade de açúcar no sangue. Em um paciente com diabete, existe uma baixa produção da proteína que ocasiona acúmulo de glicose no sangue. Isto acaba obstruindo o "caminho" de fornecimento de elementos importantes para a sobrevivência das células humanas e elas acabam morrendo desencadeando mal funcionamento do organismo. De forma básica, assim surgem os sintomas de pacientes com diabetes. 

Os genes são, portanto, trechos do DNA responsáveis pela produção de alguma proteína no organismo humano e um deles é responsável pela produção de insulina.

## Influência dos genes uns nos outros

balanca_classica.jpg

Fonte: [istockphoto](https://www.istockphoto.com/br/vetor/vetor-cl%C3%A1ssica-balan%C3%A7a-da-justi%C3%A7a-gm137911674-19037605)

Porém, as expressões gênicas não são independentes umas das outras. O aumento de um trecho pode diminuir a quantidade produzida em outro trecho, por exemplo. Portanto, quando compostos químicos são analisados como possíveis drogas, não é olhado somente o efeito desta substância em um gene específico, mas precisa se ter em mente como a alteração nesse trecho afeta a quantidade de outro.

Por exemplo, existe uma proteína chamada hemoglobina. Ela é responsável pela circulação de oxigênio em nosso sangue. Se tivermos uma quantidade abaixo do comum desta proteína, podemos ter anemia.

Assim, ao tentarmos descobrir alguma substância ou combinação destas que aumente a produção de insulina num paciente diabético, precisamos observar se ela não ocasiona, por outro lado, a diminuição de hemoglobina. Teríamos, caso isto ocorresse, um remédio com efeitos colaterais não desejáveis.

### O que é aprendizado de máquina?

Basicamente, aprendizado de máquina são técnicas para ensinar o computador a realizar uma tarefa sem que ele tenha sido dito exatamente passo a passo como fazer isso.

### Elementos principais

Imagine a operação matemática mais básica: *2 + 2 = 4*. Nesta operação nós temos:

a) dados de entrada -> 2 e 2

b) operação a ser realizada nestes dados de entrada -> soma (+)

c) dado de saída a partir da operação realizada nestes dados -> 4

Esses são os principais elementos presentes em qualquer técnica de aprendizado de máquina, porém, claro, com operações matemáticas mais complexas.


## Como o Aprendizado de Máquina pode ajudar na descoberta de novas drogas?

A quantidade de trechos de DNA e tipo de células (saudáveis e cancerígenas, por exemplo) presentes nos seres vivos são, geralmente, enormes. Só nós, seres humanos, possuímos aproximadamente 20 mil genes. Nesse sentido, seria extremamente difícil tentar observar um exemplo como os comportamentos citados nas explicações sobre os genes e mortes celulares com tantos trechos de DNA e tipos celulares a serem comparados.

## Cluster


 


cluster.png

Em aprendizado de máquina existe uma técnica chamada de Cluster. Basicamente, as técnicas de clusterização buscam agrupar indivíduos (podem ser objetos, pessoas, ou como em nosso caso genes, células e compostos) por semelhança. Nesse sentido, indivíduos com características mais parecidas são colocados em um mesmo grupo (A) em diferenciação a outro(s) (B e C, por exemplo).

Podemos, portanto, utilizar técnicas de agrupamento em genes, morte de células por seus tipos e substâncias químicas a partir de ações parecidas em genes e células por características tanto dos compostos, quanto do experimento (tempo que o compsoto agiu, quantidade da dose etc). 

# Dados

Para explorar estas possibilidades, iremos utilizar dois conjuntos de dados disponibilizados pelo [Laboratory innovation science at Havard](https://lish.harvard.edu/) disponibilizados no [kaggle](https://www.kaggle.com/c/lish-moa). 

No caso dos dois conjuntos de dados que estamos trabalhando, cada linha representa uma cultura celular. Mas o que é cultura celular? Basicamente, é um grupo de células armazenadas em algum recipiente/ambiente em laboratório para que sejam feitos experimentos. Assim, nós temos dois grande grupos: 

1) Aqueles que de fato receberam alguma substância quimíca e 

2) grupo de controle, isto é, grupos que não receberam nenhum substância para que se tenha uma comparação do comportamento natural dos genes e da morte celular desse grupo com o primeiro.

Foram retirados das células valores de expressões genéticas, morte de celulas (por tipo, tais como saudável, cancerígenas etc) e mecanismos de ativação ativados ou não. Mecanismo de ativação nos indicam duas coisas: 

a) o efeito do composto (bloqueia, inibe, ativa etc); 

b) proteína em que o efeito se dá.

Essas medidas foram tiradas com dois tipos de doses diferentes do composto (D1 e D2) em três períodos diferentes após a aplicação do composto ou armazenamento do grupo de células de controle: 24, 48 e 72 horas. Assim, temos informações do efeito do composto por quantidade da dose e tempo após a sua aplicação.

# *Genes*

Como dito na apresentação inicial, podemos ter genes que estão interligados. Quando a expressão de um aumenta a de outro pode aumentar junto e vice versa. Vamos selecionar dois genes aleatórios e analisar esta possível ligação entre eles utilizando um gráfico chamado de gráfico de dispersão.

```python
#Semente
np.random.seed(seed=456)

#Genes com correlação
g1= 'g-0'
g2= 'g-8'

#genes aleatórios
g3= 'g-' + str(np.random.randint(0, 771))
g4= 'g-' + str(np.random.randint(0, 771))

fig, axes = plt.subplots(1, 2, sharey= True, figsize=(16, 8))
fig.suptitle('Relação entre Genes', fontsize=20)

#Correlação
axes[0].set_title(g1 + ' com ' + g2, fontsize=14)
sns.scatterplot(data=dados_experimentos, x=g1, y=g2, ax= axes[0])
axes[0].set_ylabel(g2, fontsize=14), axes[0].set_xlabel(g1)

#Não correlação
axes[1].set_title(g3 + ' com ' + g4, fontsize=14)
sns.scatterplot(data=dados_experimentos, x=g3, y=g4, ax= axes[1])
axes[1].set_ylabel(g4), axes[1].set_xlabel(g3);
```

Na primeira figura podemos observar a relação entre o gene 0 e o gene 8. 

Cada ponto azul representa um grupo de células diferente. 

No eixo x (horizontal), temos as expressôes gênicas do grupo no gene 0; e no eixo y (vertical) temos sua quantidade no gene 8. 

Perceba que conforme há o aumento da quantidade de genes 0 a tendência é a diminuição (valores negativos) das expressões do genes 8. 

Nesse sentido, podemos dizer que há uma tendência do gene 8 diminuir conforme o aumento do gene 0 e vice versa, sendo possível levantar a hipótese de que um influencia no outro. 

Já no gráfico do lado direito há uma dispersão maior dos pontos, não sendo possível a mesma observação de qualquer relação entre os genes 613 e o gene 683.

Assim, se estivéssemos preocupados em aumentar as expressões do gene g-0, teríamos que nos atentar na diminuição de expressões no gene 8. Os dados neste caso foram anonimizados e não sabemos o que representa cada gene, mas a linha de raciocínio seria a explicada no início do notebook sobre a influência ou não de um gene em outro.



# *Células*

Outro aspecto importante de ser analisado são tipos celulas que possuem comportamentos similares em termos de sua multiplicação ou morte de acordo com o composto recebido.

Vamos selecionar o composto mais aplicado nos experimentos, selecionar 4 tipos de células de modo aleatório e analisar como se dá a sua multiplicação ou morte comparando com o grupo de controle.

*Obs: o composto identificado como 'cacb2b860' na verdade não é um composto. Este valor simplesmente identifica as culturas celulares que não receberam nenhum composto.*  

```python
#Total de experimentos com grupo de controle e com o composto mais testado
composto_mais_testado= dados_experimentos['composto'].value_counts()[:2]
composto_mais_testado
```

```python
#Semente
np.random.seed(seed=9)

#4 Células Aleatórias
c1= 'c-' + str(np.random.randint(0, 99))
c2= 'c-' + str(np.random.randint(0, 99))
c3= 'c-' + str(np.random.randint(0, 99))
c4= 'c-' + str(np.random.randint(0, 99))
print(c1, c2, c3, c4)
```

```python
fig, axes = plt.subplots(nrows=2, ncols= 2, figsize=(14, 10), sharey= True)
fig.suptitle('Comportamento das células por tipo com ou sem o composto ' + composto_mais_testado.index[1], fontsize=20)

#c1
axes[0][0].set_title(c1, fontsize=14)
sns.boxplot(y= c1, ax=axes[0][0], x='tratamento' , data=dados_experimentos.query('composto in @composto_mais_testado.index'))
axes[0][0].set_ylabel('', fontsize=14), axes[0][0].set_xlabel('')

#c2
axes[0][1].set_title(c2, fontsize=14)
sns.boxplot(y= c2, ax=axes[0][1], x='tratamento' , data=dados_experimentos.query('composto in @composto_mais_testado.index'))
axes[0][1].set_ylabel(''), axes[0][1].set_xlabel('')

#c3
axes[1][0].set_title(c3, fontsize=14)
sns.boxplot(y= c3, ax=axes[1][0], x='tratamento' , data=dados_experimentos.query('composto in @composto_mais_testado.index'))
axes[1][0].set_ylabel('', fontsize=14), axes[1][0].set_xlabel('')

#c4
axes[1][1].set_title(c4, fontsize=14)
sns.boxplot(y= c4, ax=axes[1][1], x='tratamento' , data=dados_experimentos.query('composto in @composto_mais_testado.index'));
axes[1][1].set_ylabel(''), axes[1][1].set_xlabel('');
```

As caixas no gráfico acima nos dizem o intervalo no qual os valores daquele elemento estão mais ocorrendo e se estão de fato concentrados ou não. A caixa mais alta, portanto, nos indica que o intervalo de valores está mais disperso; do contrário, mais concentrados.

Os pontinhos pretos nos dizem indivíduos que possuem valores destoantes dos demais, seja pra cima ou pra baixo.

As caixas azuis, neste caso, nos indicam a morte dos grupos de células (por tipo, c-92, c-54 etc) que receberam o composto '87d714366 '. As caixas laranjas aqueles que não, ou seja, comportamento das células em condições normais.


Podemos observar que o composto causou a morte das células naquelas que estão sendo analisadas, já que as caixas azuis estão situadas abaixo e em valores negativos das caixas laranjas. 

As células de tipo 92 tiveram um padrão maior nessas mortes, tendo seus valores concentrados mais ou menos entre -7 e -10. Já as células de tipo 22 tiveram um comportamento mais aleatório, apesar de também terem morrido ao receberem o composto, já que seus valores de morte se concentram, aproximadamente, entre -2 e -8.
